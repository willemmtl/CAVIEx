```julia
using Distributions, Gadfly
```

```julia
N = 1000;
dist = GeneralizedExtremeValue(10.0, 1.0, -0.5);
sample = rand(dist, N);
x = 4:0.1:12
density = pdf.(dist, x)

plot(
    layer(x=x, y=density, Geom.line, Theme(default_color="red")),
    layer(x=sample, Geom.histogram(density=true), Theme(default_color="deepskyblue")),
    Guide.manual_color_key("Légende", ["Densité", "Échantillons"], ["red", "deepskyblue"]),
    Theme(background_color="white"),
    Guide.title("Simulation de données extrêmes (mu = 10, sigma = 1, xi = -0.5)"),
    Guide.xlabel("Hauteur de pluie (mm)"),
    Guide.ylabel("Occurence"),
)
```
